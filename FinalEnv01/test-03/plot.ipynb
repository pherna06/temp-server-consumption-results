```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def get_file_data(csv, pos):
    csvf = open(csv, 'r')
    csvflines = csvf.readlines()
    csvf.close()
    
    data = []    
    for line in csvflines[1:]:
        values =  line.strip('\n[]').split(',')
        data.append(float(values[pos]))
        
    return data
```

```python
###################
### POWER GRAPH ###
###################
key = 'power'
clr = 'blue'
pos = 8
X = range(0, 201)
Xticks = np.arange(0, 201, 10)
Yticks = np.arange(35, 101, 5)

chkpt = 5

fig = plt.figure(figsize=(8,6), dpi=150)
ax = fig.add_subplot(111)

for it in range(15):
    data = get_file_data(f'checkpoint-{chkpt}/iter-{it}.csv', pos)
    ax.plot(
        X, data, 
        color = clr,
        linestyle = '-',
        linewidth = 0.75,
        marker = '',
        alpha = 0.25
    )
    
ax.axhline(
    y = 97.0, 
    xmin = 10 / 220, xmax = 1 - (10 / 220),
    color = 'cyan', 
    linestyle = '-',
    linewidth = 1.5,
)

ax.set_xlabel('Número de paso')
ax.set_xticks(Xticks)
ax.set_xticklabels(Xticks, rotation=45)

ax.set_ylabel('Potencia (w)')
ax.set_yticks(Yticks)

ax.grid(
    alpha=0.25,
    linestyle=':',
)

plt.savefig(f'img/chkpt-{chkpt}.{key}.png', bbox_inches='tight')
```

```python
###################
### STATE GRAPH ###
###################
key = 'state'
clr = 'red'
pos = 1
X = range(0, 201)
Xticks = np.arange(0, 201, 10)
Yticks = np.arange(10, 31, 2)

chkpt = 5

fig = plt.figure(figsize=(8,6), dpi=150)
ax = fig.add_subplot(111)

for it in range(15):
    data = get_file_data(f'checkpoint-{chkpt}/iter-{it}.csv', pos)
    ax.plot(
        X, data, 
        color = clr,
        linestyle = '-',
        linewidth = 0.75,
        marker = '',
        alpha = 0.25
    )
    
ax.axhline(
    y = 29, 
    xmin = 10 / 220, xmax = 1 - (10 / 220),
    color = 'purple', 
    linestyle = '-',
    linewidth = 1.5,
)

ax.set_xlabel('Número de paso')
ax.set_xticks(Xticks)
ax.set_xticklabels(Xticks, rotation=45)

ax.set_ylabel('Estado')
ax.set_yticks(Yticks)

ax.grid(
    alpha=0.25,
    linestyle=':',
)

plt.savefig(f'img/chkpt-{chkpt}.{key}.png', bbox_inches='tight')
```
