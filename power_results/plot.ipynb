```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def get_file_data(csv):
    csvf = open(csv, 'r')
    csvflines = csvf.readlines()
    csvf.close()
    
    X = []
    Y = []
    
    for line in csvflines[1:]:
        freq, power, _ = line.split(',')
        X.append( int(freq) // 1000 )
        Y.append( float(power) )
        
    return X, Y
```

```python
####################
### FLOATPRODUCT ###
####################
X, Y = get_file_data('floatproduct.power.csv')
Xticks = X
Yticks = np.arange(35, 101, 5)

fig = plt.figure(figsize=(8,6), dpi=150)

ax = fig.add_subplot(111)

ax.set_title('floatproduct')
ax.plot(
    X, Y, 
    color = 'blue', 
    label = 'floatproduct',
    linestyle = '-',
    marker = 'o'
)

ax.set_xlabel('Frecuencia (MHz)')
ax.set_xticks(Xticks)
ax.set_xticklabels(Xticks, rotation=45)

ax.set_ylabel('Potencia (w)')
ax.set_yticks(Yticks)

ax.grid(
    alpha=0.50,
    linestyle=':',
)

plt.savefig('img/floatproduct.power.png', bbox_inches='tight')
```
